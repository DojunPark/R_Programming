for loop

```r
sum <- 0
for (i in 1:10){
    sum <- sum + i
    print(sum)
}
```

while loop

```r
sum <- 0
i <- 1

while (i<=10){
    sum <- sum + i
    print(sum)
    
    i <- i + 1
}
```

repeat loop

```r
sum <- 0
i <- 1

repeat{
    sum <- sum + i
    print(sum)
    i <- i + 1
    
    if (i>10)
        break
}
```

# 6.3 Random barplot

* runif() 함수
: r uniform distribution의 약자

```r
runif(1, 0, 1)   # 0~1 사이의 난수를 1개 생성
```

```r
runif(2, 0, 1)  # 0~1 사이의 난수를 2개 생성
```

```r
runif(3, -5, 5)  # 음수도 영역으로 지정 가능
```

- 1초 간격으로 그래프 그리기

```r
install.packages('animation')
library(animation)
```

```r
ani.options(interval = 1)

# 무한반복
while(TRUE){
    y <- runif(5, 0, 1)
    barplot(y, ylim=c(0, 1), col=rainbow(5))
    ani.pause()
}
```

```r
library(magick)

#이미지 설정
space_bg <- image_read('c://R/image/space.png')
rocket_1 <- image_read('c://R/image/rocket1.png')
rocket_2 <- image_read('c://R/image/rocket2.png')

space_bg <- image_scale(space_bg, '600x600!')
rocket_1 <- image_scale(rocket_1, '200x200!')
rocket_2 <- image_scale(rocket_2, '200x200!')

#애니메이션 설정 및 실행
moving <- 0
x <- 200
y <- 600

while(TRUE){
    position <- paste('+', x, '+', y, sep='')

    if(moving%%2 == 0){
        img <- image_composite(space_bg, rocket_1, offset=position)
    }else{
        img <- image_composite(space_bg, rocket_2, offset=position)
    }

    plot(img)
    Sys.sleep(0.3)

    if(y <= 0)
        break

    moving <- moving + 1
    y <- y - 100
}
```
