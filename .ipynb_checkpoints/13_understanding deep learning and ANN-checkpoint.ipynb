## step2. 분석 대상 데이터 할당

```r
data <- Boston[, c(1, 2, 4, 5, 9, 14)]  # 6개의 feature 선택 (독립변수 5개, 종속변수 1개)

head(data)
```

## step3. 결측치 확인

```r
na <- apply(data, 2, is.na)  # 열 기준으로(2) is.na함수를 적용

head(na)
```

```r
apply(na, 2, sum)   # 열별 결측치 누적 갯수 확인
```

## step4. 데이터 정규화

```r
maxs <- apply(data, 2, max)   # 각 열의 최대값을 구함
maxs
```

```r
mins <- apply(data, 2, min)  # 각 열의 최소값을 구함
mins
```

```r
data.scaled <- scale(data, center=mins, scale=maxs-mins)  # 데이터에서 mins 만큼 빼고, maxs-mins 만큼 나눔

head(data.scaled)
```

## step5. 학습, 테스트 데이터 분리

```r
n <- nrow(data.scaled)
n
```

```r
set.seed(1234)
index <- sample(1:n, round(0.8*n))  # 1부터 n(전체데이터수)까지의 수에서 80%의 인덱스 번호를 랜덤으로 뽑음

index
```

```r
train <- as.data.frame(data.scaled[index,])  # 표본추출된 인덱스번호의 row를 데이터프레임으로 저장
                                            # 사실 원래 데이터프레임 데이터였기에  as.data.frame() 함수를 쓰지 않아도 무방

print(dim(train))
head(train)
```

```r
test <- as.data.frame(data.scaled[-index,])  # 인덱스번호의 row를 빼고 저장하면 그게 테스트 데이터

print(dim(test))
head(test)
```

## step6. 학습 데이터 입, 출력 변수 할당

```r
names.col <- colnames(train)
names.col
```

```r
var.dependent <- names.col[6]
var.dependent   # 종속변수 (출력 변수)
```

```r
var.independent <- names.col[-6]
var.independent  # 독립변수 (입력 변수)
```

```r
f.var.independent <- paste(var.independent, collapse=" + ")  # 입력 변수를 +로 연결
f.var.independent
```

```r
f <- paste(var.dependent, '~', f.var.independent)
f
```

## step7. 모델 학습 및 그래프 출력

```r
model <- neuralnet(f, data=train, hidden=c(3,2), linear.output = T)

comment1 <- "
    f: 입출력 변수를 의미
    data: 학습 데이터
    hidden: 은닉층의 노드 수 (제1은닉층 3개, 제2은닉층 2개)
    linear.out: 출력 노드의 활성화 함수 적용 여부
    * 은닉층의 활성화함수가 따로 지정되지 않으면 시그모이드 함수가 적용됨
"

plot(model)  # jupyter notebook에서는 출력되지 않음
```

# step8. 테스트

```r
predicted <- compute(model, test[, var.independent])
```

```r
head(predicted$net.result)
```

```r
predicted.real <- predicted$net.result * (maxs[5] - mins[5]) + mins[5]  # 교재에 설명된 부분 (?)
head(predicted.real)
```

```r
predicted.real <- predicted$net.result * (maxs[6] - mins[6]) + mins[6]  # 원래 스케일로 복원
head(predicted.real)
```

```r
test.real <- test$medv * (maxs[6] - mins[6]) + mins[6]
test.real
```

## step9. 실제값 대비 예측 값 비교

```r
plot(test.real, predicted.real,
    xlim=c(0, 50), ylim=c(0, 50),
    main='실제 값 대비 예측 값 분포',
    xlab='실제 값', ylab='예측 값',
    col='red', pch=18, cex=0.7)
abline(0, 1, col='blue', lty=2)
```

```r
MAPE.nn <- sum(abs(test.real-predicted.real)/test.real*100)/length(test.real)
MAPE.nn
```

```r
MAPE.nn <- sum(abs(test.real-predicted.real)/test.real*100)/nrow(as.data.frame(test.real))
MAPE.nn
```
