# 5.3. Pie chart

```r
city <- c("서울", '부산', '대구', '인천', '광주', '대전', '울산')
pm25 <- c(18, 21, 21, 17, 8, 11, 25)
colors <- c('red', 'orange', 'yellow', 'green', 'lightblue','blue', 'violet')

pie(pm25, labels = city, col=colors, main='지역별 초미세먼지 농도')
```

추가 옵션을 주어서 90도에서 시작하여 시계 방향으로 회전

```r
pie(pm25, labels=city, col=colors, main='지역별 초미세먼지 농도', init.angle=90, clockwise=T)
```

gdRdriver 패키지의 팔레트 - 무지개색

라벨 변경과 팔레트 사용

```r
greens <- brewer.pal(7, 'Greens')
pct <- round(pm25/sum(pm25)*100, 0)  # 소수점 0번째까지 입력 (반올림하여 정수로 입력)
city_label <- paste(city, ', ', pct, '%', sep='')

pie(pm25, labels=city_label, col=greens, main='지역별 초미세먼지 농도', init.angle=90, clockwise=T)
```

# 5.4. Bar chart

```r
dept <- c('영업 1팀', '영업 2팀', '영업 3팀', '영업 4팀')
sales01 <- c(4, 12, 5, 8)

barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
```

수직선 출력

```r
barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
abline(h=mean(sales01), col='orange', lty=2)
```

데이터 라벨 출력

```r
bp <- barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
              xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
text(x=bp, y=sales01, labels=sales01, pos=3)
```

수평으로 출력

```r
barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15), horiz=T)
```

스택형 바 차트

```r
dept <- c('영업 1팀', '영업 2팀', '영업 3팀', '영업 4팀')
sales01 <- c(4, 12, 5, 8)
sales02 <- c(12, 8, 8, 4)
sales <- rbind(sales01, sales02)
sales
```

```r
legend_lbl <- c('1분기', '2분기')
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='top',  ncol=2))
```

```r
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='right'))
```

그룹형 바 차트

```r
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='right',  ncol=2),
       beside=T)
```

# Histogram

```r
quakes
```

```r
mag <- quakes$mag
mag
```

```r
hist(mag)
```

```r
hist(mag, breaks=5)  # 계급 구간의 수 설정
```

```r
hist(mag, breaks=10)
```

```r
hist(mag,breaks=seq(4, 6.5, by=0.5))  # 4~6.5까지 0.5간격
```

```r
hist(mag, breaks=seq(4, 6.5, by=0.5),
    col=rainbow(5))
```

```r
hist(mag, breaks=seq(4, 6.5, by=0.5),
    col=rainbow(5),
    freq=F)          # 상대빈도로 출력
```

```r
hist(mag, breaks=seq(4, 6.5, by=0.5),
    col=rainbow(5),
    probability=T)   # 상대 빈도로 출력 (freq=F와 동일)
```

- breaks="Sturges" 로 설정하면 계급의 수를 $\sqrt{n}$으로 자동 계산함

```r
hist(mag, breaks='Sturges',
    col=rainbow(5),
    freq=T)
```

# 5.7 Box Plot

```r
mag <- quakes$mag
mag
```

```r
boxplot(mag)
```

```r
boxplot(mag,
       main='지진 발생 강도의 분포',
       xlab='지진', ylab='지진 규모',
       col='red')
```

```r
summary(mag)
```

```r
quantile(mag)  # summary에서 mean이 빠진 것이 차이
```

```r
min(mag)
```

```r
max(mag)
```

```r
q <- quantile(mag)
q[1]
```

```r
iqr <- q[4] - q[2]
iqr
```

```r
IQR(mag)   # 위에 직접 계산한 식과 동일함
```

```r
fence.upper <- q[4] + 1.5*IQR(mag)   # 상위기준
fence.upper
```

```r
fence.lower <- q[2] - 1.5*IQR(mag)   # 하위 기준
fence.lower
```

- 이상치 출력

```r
mag[mag > fence.upper]  # 상위 기준 초과 이상치  (7개)
```

```r
mag[mag < fence.lower]  # 하위 기준 미만 이상치  (0개)
```

```r
min(mag[mag > fence.upper])  # 이상치 중 최소값
```

```r
fence.upper   # 상위기준
```

```r
max(mag[mag < fence.upper])  # 상위기준 내 최대값
```

```r
fence.lower  # 하위 기준
```

```r
min(mag[mag>fence.lower])  # 하위 기준 내 최소값
```
