# Get posts with the keyword "여행"

```r
search_url <- 'https://openapi.naver.com/v1/search/blog.xml'
query <- URLencode(iconv('여행', 'euc-kr', 'UTF-8'))
url <- paste(search_url, '?query=', query, '&display=20', sep='')
```

```r
url
```

```r
doc <- getURL(url,
             httpheader = c('Content-Type' = 'application/xml',
                           'X-Naver-Client-Id' = client_id,
                           'X-Naver-Client-Secret' = client_secret))
```

```r
doc
```

# extract texts from xml

```r
xmlFile <- xmlParse(doc)
df <- xmlToDataFrame(getNodeSet(xmlFile, '//item'))
str(df)
```

```r
xmlRoot(xmlFile)
```

```r
xmlToDataFrame(getNodeSet(xmlFile, '//item'))
```

```r
description <- df[, 3]
print(description)
```

```r
print(description[1])
```

# Text preprocessing

```r
description2 <- gsub('<b>|</b>|&quot;|#|▲|*|♡', '', description)
print(description2)
```

# Noun extraction

```r
nouns <- nouns(iconv(description2, 'utf-8'))
print(nouns)
```

# Make vocab list

```r
nouns.all <- unlist(nouns, use.name=F)
print(nouns.all)
```

```r
# 단어길이가 1인 리스트
nouns.all1 <- nouns.all[nchar(nouns.all) <= 1]

# 단어길이가 2 이상인 단어 리스트
nouns.all2 <- nouns.all[nchar(nouns.all) >= 2]
```

```r
nouns.all1
```

```r
nouns.all2
```

# Make count list

```r
nouns.freq <- table(nouns.all2)
nouns.freq
```

```r
nouns.df <- data.frame(nouns.freq, stringsAsFactors=F)
nouns.df
```

# Sorting

```r
nouns.df.sort <- nouns.df[order(-nouns.df$Freq),]  # 내림차순 정렬
nouns.df.sort
```

# Delete the keyword "여행"

```r
myData <- nouns.df.sort[-c(1), ]
myData
```

# Wordcloud

```r
wordcloud(myData[,1],
         myData[,2],
         min.freq=1,
         scale=c(3, 0.5),
         rot.per=0.25,
         random.order=F,
         random.color=T,
         colors=rainbow(10))
```
