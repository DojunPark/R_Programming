# 5.3. Pie chart

```r
city <- c("서울", '부산', '대구', '인천', '광주', '대전', '울산')
pm25 <- c(18, 21, 21, 17, 8, 11, 25)
colors <- c('red', 'orange', 'yellow', 'green', 'lightblue','blue', 'violet')

pie(pm25, labels = city, col=colors, main='지역별 초미세먼지 농도')
```

추가 옵션을 주어서 90도에서 시작하여 시계 방향으로 회전

```r
pie(pm25, labels=city, col=colors, main='지역별 초미세먼지 농도', init.angle=90, clockwise=T)
```

라벨 변경과 팔레트 사용

```r
greens <- brewer.pal(7, 'Greens')
pct <- round(pm25/sum(pm25)*100, 0)  # 소수점 0번째까지 입력 (반올림하여 정수로 입력)
city_label <- paste(city, ', ', pct, '%', sep='')

pie(pm25, labels=city_label, col=greens, main='지역별 초미세먼지 농도', init.angle=90, clockwise=T)
```

# 5.4. Bar chart

```r
dept <- c('영업 1팀', '영업 2팀', '영업 3팀', '영업 4팀')
sales01 <- c(4, 12, 5, 8)

barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
```

수직선 출력

```r
barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
abline(h=mean(sales01), col='orange', lty=2)
```

데이터 라벨 출력

```r
bp <- barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
              xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15))
text(x=bp, y=sales01, labels=sales01, pos=3)
```

수평으로 출력

```r
barplot(sales01, names.arg = dept, main='부서별 영업 실적(1분기)', col=rainbow(length(dept)), 
        xlab='부서', ylab='영업 실적(억원)', ylim=c(0, 15), horiz=T)
```

스택형 바 차트

```r
dept <- c('영업 1팀', '영업 2팀', '영업 3팀', '영업 4팀')
sales01 <- c(4, 12, 5, 8)
sales02 <- c(12, 8, 8, 4)
sales <- rbind(sales01, sales02)
sales
```

```r
legend_lbl <- c('1분기', '2분기')
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='top',  ncol=2))
```

```r
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='right'))
```

그룹형 바 차트

```r
barplot(sales, main='부서별 영업 실적(1~2분기)', names.arg=dept,
       col=c('green', 'orange'),
       legend.text=legend_lbl, ylim=c(0, 30),
       args.legend=list(x='right',  ncol=2),
       beside=T)
```

```r
height <- women$height
weight <- women$weight

plot(x=height, y=weight, xlab='키', ylab='몸무게', main='키와 몸무게의 변화')
```

```r
plot(x=height, y=weight, xlim=c(min(height), max(height)), ylim=c(min(weight)-10, max(weight)+10),
     xlab='키 (인치)', ylab='몸무게 (파운드)', main='키와 몸무게의 변화',
    col='blue', bg='green', pch=21, cex=3, type='b', lty=4, lwd=2)
```
